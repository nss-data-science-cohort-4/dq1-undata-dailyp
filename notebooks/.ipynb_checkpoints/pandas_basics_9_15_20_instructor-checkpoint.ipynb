## Step 3: Bivariate Analysis/Relationships Between Variables

```python
house_sales.info()
```

```python
house_sales[
    (house_sales['yr_renovated']>0)
    &(house_sales['price']<1000000)
].plot(kind='scatter', x='yr_renovated', y='price')
```

```python
house_sales['waterfront'].value_counts()
```

```python
house_sales.head()
```

```python
house_sales.groupby('date')['id'].count().plot(figsize=(15,8))
```

```python
house_sales[
    (house_sales['date'].astype("datetime64[ns]") >= '2014-05-01')
    &(house_sales['date'].astype("datetime64[ns]") < '2014-06-01')
].groupby('date')['id'].count().plot(figsize=(15,8))
```
